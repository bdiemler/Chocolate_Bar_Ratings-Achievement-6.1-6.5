# Renaming columns 

```python
# Renaming 
df_choc.rename(columns = {'Company\xa0\n(Maker-if known)' : 'company_manufacturer'}, inplace = True)
df_choc.rename(columns = {'Specific Bean Origin\nor Bar Name' : 'bar_origin'}, inplace = True)
df_choc.rename(columns = {'REF' : 'review_ID'}, inplace = True)
df_choc.rename(columns = {'Review\nDate' : 'review_date'}, inplace = True)
df_choc.rename(columns = {'Cocoa\nPercent' : 'cocoa_percentage'}, inplace = True)
df_choc.rename(columns = {'Company\nLocation' : 'company_location'}, inplace = True)
df_choc.rename(columns = {'Rating' : 'rating'}, inplace = True)
df_choc.rename(columns = {'Bean\nType' : 'bean_type'}, inplace = True)
df_choc.rename(columns = {'Broad Bean\nOrigin' : 'bean_region'}, inplace = True)
```

# Missing Values 

```python
# Finding Missing Values 
df_choc.isnull().sum()
```

```python
# Identifying the bean_type string missing
df_choc['bean_type'].unique()
```

```python
# Replacing '\xa0' with NaN
df_choc['bean_type'] = df_choc['bean_type'].replace('\xa0', np.nan)
```

```python
# Rerun the missing values check to see the change
df_choc.isnull().sum()
```

```python
# Finding Missing bean_type in cells
df_nan = df_choc[df_choc['bean_type'].isnull() == True]
```

```python
df_nan
```

```python
# Finding Missing bean_region in cells
df_nan_region = df_choc[df_choc['bean_region'].isnull() == True]
```

```python
df_nan_region
```

# Dropping columns for bean_type

```python
# Dropping bean_type as almost half of the data is missing - imputation could introduce significant bias
df = df_choc.drop('bean_type', axis=1)
```

```python
df.shape
```

```python
df.head(5)
```

# Imputation for bean_region

```python
# Analysing previous data, many bar_origin are similar to bean_region. Filtering bar_origin = Madagascar
df_madagascar = df_choc[df_choc['bar_origin'] == 'Madagascar']
```

```python
df_madagascar ['bean_region'].mode()
```

```python
df_madagascar
```

```python
# Replacing NaN value with 'Madagascar' using contextual clue observed above
df['bean_region'] = df_choc['bean_region'].fillna('Madagascar')
```

```python
# Re-checking nulls
df.isnull().sum()
```

# Duplicates

```python
# Finding Duplicates
df_dups = df[df.duplicated()]
```

```python
# Checking Duplicates
df_dups
```

# Converting percentage strings on decimal floats

```python
# 1. Make sure it is a string
df['cocoa_percentage'] = df['cocoa_percentage'].astype(str)

# 2. Remove the '%' sign and strip any extra spaces
df['cocoa_percentage'] = df['cocoa_percentage'].str.replace('%', '').str.strip()

# 3. Convert the cleaned string to a float and divide by 100
df['cocoa_percentage'] = df['cocoa_percentage'].astype(float) / 100
```

```python
# Checking the first few values to ensure the conversion worked
print(df['cocoa_percentage'].head())
```
